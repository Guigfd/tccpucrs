# Inspecionando o Dataframe

## Descrição de cada coluna do dataframe:

* Rank - O rank do jogo de acordo com as vendas gerais
* Name - Nome do jogo
* Platform - Plataforma que o jogo foi lançado (i.e PC, PS4, etc)
* Year_of_Release - Ano que o jogo foi lançado
* Genre - gênero do jogo
* Publisher - A distribuidora do jogo
* NA_Sales - Vendas na América do Norte (em milhões de cópias)
* EU_Sales - Vendas na Europa ((em milhões de cópias)
* JP_Sales - Vendas no Japão (em milhões de cópias)
* Other Sales - Vendas no resto do mundo (em milhões de cópias)
* Global_Sales - Total de vendas ao redor do mundo
* Critic_Score - Pontuação agregada compilada pela equipe do Metacritic
* Criticcount - O número de críticos para criar o critic score
* User_score - Pontuação dos assinantes do Metacritic
* Usercount - Número de usuários que deram a pontuação de usuário
* Developer - Desenvolvedor do jogo
* Rating - Classificação etária do jogo

Informações tiradas do próprio criador do dataset (https://www.kaggle.com/datasets/rush4ratio/video-game-sales-with-ratings)
